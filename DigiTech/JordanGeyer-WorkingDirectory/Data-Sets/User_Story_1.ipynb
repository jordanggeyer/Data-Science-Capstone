```python
#Getting rid of the decimals via string so we can ensure we can convert to MB correctly
appData['Size'] = appData['Size'].str.replace(".0", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".1", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".2", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".3", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".4", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".5", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".6", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".7", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".8", "", regex=False)
appData['Size'] = appData['Size'].str.replace(".9", "", regex=False)
appData['Size'] = appData['Size'].str.replace("k", "000", regex=False) #Multiplying by the Unit Conversion
appData['Size'] = appData['Size'].str.replace("M", "000000", regex=False) #Multiplying by the Unit Conversion
appData['Size'] = appData['Size'].str.replace("Varies with device", "", regex=False) #Want to remove this as the data is not discrete
appData['Size'] = pd.to_numeric(appData['Size'], errors='coerce') #Coerces all values to be moved to the numerical data type
appData['Size'] = appData['Size'].fillna(0) #Take care of any blanks left
appData['Size'] = appData['Size'].astype(int) #Convert from float to int to drop the decimal
appData['Size'] = appData['Size'] / 1000 #Divided by the conversion amount
appData.rename(columns={'Size': 'KB'}, inplace=True) #Shows the amount of storage usage - KB
```

```python
#Remove symbols "$" that will prevent from data type change
appData['Price'] = appData['Price'].str.replace('$', '', regex=False)

#Converting from object to float
appData['Price'] = appData['Price'].astype(float)
```

Data Analysis

```python
appClean = pd.read_csv("updated_appData.csv")  # Reading the updated dataset in the csv
```

```python
#Dropping Columns not necessary for analysis
appClean.drop(columns=['App'], inplace=True)
appClean.drop(columns=['Last Updated Formatted'], inplace=True)
```

```python
# Select all rows and all columns starting from index 1
X = appClean.iloc[:, 1:]

# Set y to the "Categories" column
y = appClean.iloc[:, 0].values
```

```python
#Encoding the "Category" column to numerical
y_encoded = pd.get_dummies(y, drop_first=True)

# Display the encoded DataFrame to ensure success
print(y_encoded.head())
```

```python
# Combine X and y_encoded
combined_data = pd.concat([X.reset_index(drop=True), y_encoded.reset_index(drop=True)], axis=1)

# Display the first few rows of the combined DataFrame
print(combined_data.head())
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Display the shapes of the splits
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#Fitting Multiple Linear Regression to the Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting the Test set Results
y_pred = regressor.predict(X_test)
y_pred
```

```python
#Calculating the Coefficients
print(regressor.coef_)
```

```python
#Calculating the Intercept
print(regressor.intercept_)
```

```python
#Calculating the R Scored Value
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
